```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from tslearn.clustering import TimeSeriesKMeans, KShape
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load and Preprocess the Data
# Load the dataset
file_path = "India_Condom_Market_Dataset.csv"
data = pd.read_csv(file_path)

# Handle missing values
data.dropna(inplace=True)

# Convert 'Year' to datetime format
data['Year'] = pd.to_datetime(data['Year'], format='%Y')

# Select relevant columns for clustering
columns = ['Year', 'Market Size (USD Million)', 'CAGR (%)', 'Material Type', 'Product Type', 'Distribution Channel', 'Region']
time_series_data = data[columns]

# Step 2: Apply Time-Series Clustering
# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(time_series_data[['Market Size (USD Million)', 'CAGR (%)']])

# Reshape data for time-series clustering
reshaped_data = scaled_data.reshape((scaled_data.shape[0], 1, scaled_data.shape[1]))

# Apply K-means clustering
kmeans = TimeSeriesKMeans(n_clusters=5, metric="dtw", max_iter=10, random_state=42)
kmeans_labels = kmeans.fit_predict(reshaped_data)

# Step 3: Analyze Clusters
# Add cluster labels to the original data
data['Cluster'] = kmeans_labels

# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Market Size (USD Million)', hue='Cluster', data=data, palette='tab10')
plt.title('Time-Series Clustering of Condom Sales Trends')
plt.xlabel('Year')
plt.ylabel('Market Size (USD Million)')
plt.legend(title='Cluster')
plt.show()
```
